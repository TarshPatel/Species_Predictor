## Species_Predictor

Aim: To predict the Species from new traits of a flower.

### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading dataset

```python
dataset = pd.read_csv(r"C:\Desktop\Quick Access\Iris.csv")
x = dataset.iloc[:,1:-1]
y = dataset.iloc[:,-1].values
```

### Splitting the data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25,random_state = 0)
```

### Training the model

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy',random_state = 0)
classifier.fit(x_train,y_train)
```

### Comparing the predicted species with the actual species

```python
y_pred = classifier.predict(x_test)
print ('Predicted Species   Actual Species')
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(x_test),1)),1))
```

### Predicting from new traits of the flower

```python
new_trait = [[0.2,23,5,1]]
classifier.predict(new_trait)
```

### Evaluating the result of the model

```python
from sklearn.metrics import accuracy_score,confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

### Creating the decision tree

```python
features = ('Sepal Length','Sepal Width','Petal Length','Petal Width')
```

```python
from sklearn.tree import plot_tree
dtree = plt.figure(figsize = (20,15))
dtree = plot_tree(classifier,feature_names=features,class_names = dataset['Species'].unique(),rounded = True,filled = True,fontsize = 15)
```

### Conclusion

The above tree can be used to predict the species of the flower with given traits.
